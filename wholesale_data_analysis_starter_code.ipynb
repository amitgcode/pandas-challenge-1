```python
# What three item categories had the most entries?
count_category_df = df.groupby("category").agg(category_count = ("category", "count"))
top_category_df_count = count_category_df.sort_values("category_count", ascending=False).head(1)
top_category_df_count
```

```python
# For the category with the most entries, which subcategory had the most entries?
top_category_item_df =  df.join(top_category_df_count, how="inner", on="category")
top_category_item_df.groupby(["category","subcategory"]).agg(subcategory_count = ("subcategory", "count")).sort_values("subcategory_count", ascending=False).head(1)
```

```python
# Which five clients had the most entries in the data?
df_top5_clients = df.groupby("client_id").agg(entry_count = ("client_id", "count")).sort_values("entry_count", ascending=False).head(5)
df_top5_clients
```

```python
# Store the client ids of those top 5 clients in a list.
lst_clientid = [x for x in df_top5_clients.index]
lst_clientid
```

```python
# How many total units (the qty column) did the client with the most entries order order?
top_client_item_df = df.loc[df["client_id"]==lst_clientid[0],:]
order_qty_df =  top_client_item_df.groupby("order_id").agg(nos_of_order = ("order_id","count"), total_units=("qty","sum"))
order_qty_df.sort_values(["nos_of_order","total_units"], ascending=False,inplace=True)
order_qty_df.head(1)
```

## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
#df.head(5)
client_specific_df = df.copy()
client_specific_df["subtotal"] = client_specific_df["unit_price"] * client_specific_df["qty"]
client_specific_df.head(5)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
client_specific_df["total_weight"] = client_specific_df["unit_weight"] * client_specific_df["qty"]
client_specific_df["shipping_price"] = round( client_specific_df["total_weight"].apply( lambda weight: (weight * 7.00) if  weight > 50.00 else weight * 10.00 ) ,2)
client_specific_df.head(5)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
client_specific_df["total_tax"] = round(((client_specific_df["subtotal"] + client_specific_df["shipping_price"])*9.25/100),2)
client_specific_df["total_price"] = round(client_specific_df["subtotal"] + client_specific_df["shipping_price"] + client_specific_df["total_tax"],2) 
client_specific_df.head(5)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
client_specific_df["line_cost"]  = round((client_specific_df["unit_cost"] * client_specific_df["qty"]) + client_specific_df["shipping_price"],2)
client_specific_df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price
client_specific_df["profit"]  = client_specific_df["total_price"] - client_specific_df["line_cost"]
client_specific_df.head(5)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

df_clientid = pd.DataFrame(lst_clientid, columns=["client_id"])

top_client5_item_df = pd.merge(client_specific_df,df_clientid, on="client_id")
top_client5_total_qty_df = top_client5_item_df.groupby("client_id").agg(total_quantity = ("qty","sum"))
top_client5_total_qty_df.head(5)
top_client5_item_df.info()
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
top_client5_summarize = top_client5_item_df.groupby("client_id").agg(total_units_purchased = ('qty','sum'),
        total_shipping_price=('shipping_price','sum'),total_revenue=('subtotal','sum'),total_profit=('profit','sum'))

top_client5_summarize.head(10)
```

```python
# Format the data and rename the columns to names suitable for presentation.


# Define the money columns. 
top_client5_summarize.rename(columns={"total_units_purchased":"Total Units Purchased", 
                                      "total_revenue":"Total Revenue",
                                      "total_shipping_price":"Total Shipping Price",
                                      "total_profit":"Total Profit",
                                      }, inplace=True)

top_client5_summarize.head()
# Define a function that converts a dollar amount to millions.
def currency_format_millions(x):
    return round(x/1000000,2)

# Apply the currency_format_millions function to only the money columns. 

top_client5_summarize["Total Revenue $(millions)"] = top_client5_summarize["Total Revenue"].apply(currency_format_millions) 
top_client5_summarize["Total Shipping Cost $(millions)"] = top_client5_summarize["Total Shipping Price"].apply(currency_format_millions) 
top_client5_summarize["Total Profit $(millions)"] = top_client5_summarize["Total Profit"].apply(currency_format_millions) 

final_client5_summary = top_client5_summarize[["Total Units Purchased", "Total Shipping Cost $(millions)","Total Revenue $(millions)","Total Profit $(millions)"]]
# Rename the columns to reflect the change in the money format. 
final_client5_summary.head(10)

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
final_client5_summary_sorted = final_client5_summary.sort_values("Total Profit $(millions)", ascending=False)
final_client5_summary_sorted.head()
```
